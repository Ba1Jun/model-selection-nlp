```python
tuned_results = {
    "mnli": [84.7, 82.2, 83.9, 81.5, 87.6, 84.0],
    "qqp": [89.6, 88.5, 89.2, 87.8, 91.9, 89.4],
    "qnli": [91.8, 89.2, 91.0, 88.2, 92.8, 90.8],
    "sst2": [92.7, 91.3, 91.7, 90.4, 94.8, 92.5],
    "cola": [56.3, 51.3, 58.2, 47.2, 63.6, 59.3],
    "mrpc": [88.6, 87.5, 87.8, 85.6, 90.2, 86.6],
    "rte": [69.3, 59.9, 66.1, 60.6, 78.7, 67.9],
    "wnli": [53.5, 56.3, 46.5, 56.3, 57.7, 52.1],
}

dataset_sizes = {
    "mnli": 392702,
    "qqp": 363846,
    "qnli": 104743,
    "sst2": 67349,
    "cola": 8551,
    "mrpc": 3668,
    "rte": 2490,
    "wnli": 635,
}

datasets = ['cola', 'sst2', 'mrpc', 'qqp', 'mnli', 'qnli', 'rte', 'wnli']
encoders = ["bert-base-uncased",
            "distilbert-base-uncased",
            "bert-base-cased",
            "distilbert-base-cased",
            "roberta-base",
            "distilroberta-base"]

all_corr_results = {}

pooling = 'first'
target_model_lm = "albert-base-v2"
model_sim_methods = ["DSE", "PA", "RSA", "DDS"]
static_methods = ["kNN-3", "GBC", "HScoreR", "MSC", "LFC",  "PARC", "TransRate", "NLEEP"]
dynamic_methods = ["Logistic", "LogME", "SFDA", "PACTran-1-100"]
training_methods = []

all_methods = model_sim_methods + static_methods + dynamic_methods


model_sim_methods = ["DSE-corr", "DSE-dot", "DSE-cos", "DSE-l2", "PA",\
                    "RSA-corr", "RSA-cos", "RSA-l2", "RSA-dot", "DDS-corr", "DDS-cos", "DDS-dot", "DDS-l2", ]


all_methods = [
               "MSC-l2", "MSC-cos", "MSC-corr", "GBC", "HScore", "HScoreR", "kNN-1-corr", "kNN-3-corr", "kNN-5-corr",\
               "kNN-1-l2", "kNN-3-l2", "kNN-5-l2", "kNN-1-cos", "kNN-3-cos", "kNN-5-cos",\
               "LFC-corr", "LFC-cos", "LFC-dot", "LFC-l2", "PA", \
               "PARC-corr", "PARC-cos", "PARC-dot", "PARC-l2", ,\
               "Logistic", "LogME", "SFDA", "PACTran-1-100", "PACTran-1-10", "PACTran-10-1000", "PACTran-10-100",\
               "PACTran-0.1-10", "PACTran-0.1-1", "TransRate", "NLEEP"] 
method_show_name = dict(zip(all_methods, all_methods))

# method_show_name = {
#     "EarlyStop-50": "EarlyStop (50 Steps)",
#     "EarlyStop-100": "EarlyStop (100 Steps)",
#     "EarlyStop-500": "EarlyStop (500 Steps)",
    
#     "DSE-MeanEmb": "DSE (MeanEmb)",
#     "DSE-MeanSim": "DSE",
#     "DSE-dot": "DSE-dot",
#     "PA": "PD", 
#     "RSA": "RSA", 
#     "DDS": "DDS",
    
#     "kNN-1": "kNN (k=1)", 
#     "kNN-3": "kNN", 
#     "kNN-5": "kNN (k=5)",
    
#     "kNN-1-l2": "kNN-1-l2",
#     "kNN-3-l2": "kNN-3-l2", 
#     "kNN-5-l2": "kNN-5-l2",
#     "kNN-1-cos": "kNN-1-cos",
#     "kNN-3-cos": "kNN-3-cos", 
#     "kNN-5-cos": "kNN-5-cos", 
    
#     "NLEEP": "$\mathcal{N}$LEEP",
#     "GBC": "GBC",
#     "HScore": "H-score",
#     "HScoreR": "Reg. H-score",
#     "TransRate": "TransRate",
#     "LFC": "LFC",
#     "MSC": "MSC",
#     "PARC": "PARC", 
    
#     "LogME": "LogME",
#     "SFDA": "SFDA", 
#     "Logistic": "Logistic",
#     "PACTran-10-1000": "PACTran-10-1000",
#     "PACTran-10-100": "PACTran-10-100",
#     "PACTran-1-100": "PACTran",
#     "PACTran-1-10": "PACTran-1-10",
#     "PACTran-0.1-10": "PACTran-0.1-10",
#     "PACTran-0.1-1": "PACTran-0.1-1",
    
# }
```

```python
def ground_truth_ranking(pred_list, gt_list):
    pred_rank = np.argsort(-np.array(pred_list))
    rank = np.where(pred_rank==np.argmax(gt_list))[0].item() + 1
    return rank

def avg_time(time_list, unit):
    sum_time = 0.
    for t in time_list:
        if t == '-':
            return t
        sum_time += float(t)
    try:
        return f"%.1f%s"% (sum_time / len(time_list), unit)
    except:
        return "-"
```

```python
def summarize_results(data_ratios, embedding_sizes):

    all_results = {}
    main_results = {}
    for data_ratio in data_ratios:
        all_results[data_ratio] = {}
        main_results[data_ratio] = {}
        for embed_size in embedding_sizes:
            all_results[data_ratio][embed_size] = {}
            main_results[data_ratio][embed_size] = {}
            for method in all_methods:
                all_results[data_ratio][embed_size][method] = defaultdict(list)
                for dataset in datasets:
                    method_result_file = f"../../resources/output/glue/{dataset}/model_selection_results/{method+'-'+target_model_lm if any([method.split('-')[0]==m for m in model_sim_methods]) else method}_{int(data_ratio*100)}%_{embed_size}_{pooling}.jsonl"
                    if not os.path.exists(method_result_file):
                        print(method_result_file)
                        break
                    
                    reader =  list(jsonlines.open(method_result_file))
                    rank = 0.
                    spr = 0.
                    pear = 0.
                    wtau = 0.
                    scoring_time = 0.
                    try:
                        num_seeds = len(reader[0]['all_scores'])
                        assert len(reader) == len(encoders)
                    except:
                        break
                    for seed_idx in range(num_seeds):
                        model_result_list = []
                        model_time_list = []
                        for enc_idx, result_dict in enumerate(reader):
                            assert encoders[enc_idx] == result_dict['model'] 
                            model_result_list.append(result_dict['all_scores'][seed_idx])
                            model_time_list.append(result_dict['all_times'][seed_idx])
                        
                        assert enc_idx == len(encoders)-1
                        
                        revert_methods = ['PACTran']
                        if any([method.startswith(m) for m in revert_methods]):
                            model_result_list = [s*-1 for s in model_result_list]
                        
                        rank += ground_truth_ranking(model_result_list, tuned_results[dataset])
                        # if np.std(model_result_list) == 0:
                        #     print(method_result_file)
                        spr += spearmanr(model_result_list, tuned_results[dataset]).statistic
                        pear += pearsonr(model_result_list, tuned_results[dataset]).statistic
                        wtau += weightedtau(model_result_list, tuned_results[dataset]).statistic
                        scoring_time += np.sum(model_time_list)
                    
                    rank /= num_seeds
                    spr /= num_seeds
                    pear /= num_seeds
                    wtau /= num_seeds
                    scoring_time /= num_seeds
                    
                    training_time = '-'
                    if any([method.split('-')[0]==m for m in model_sim_methods]):
                        target_training_file = f"../../resources/output/glue/{dataset}/encoded_dataset/target-model_{target_model_lm}_{pooling}/target-model-5_first.jsonl"
                        with jsonlines.open(target_training_file) as reader:
                            for result_dict in reader:
                                training_time = str(round(result_dict['avg_time'] / (60 * 60), 2))
                                break

                    all_results[data_ratio][embed_size][method]['acc@1'].append(float(rank==1))
                    all_results[data_ratio][embed_size][method]['mrr'].append(1./rank)
                    all_results[data_ratio][embed_size][method]['spr'].append(spr)
                    all_results[data_ratio][embed_size][method]['pear'].append(pear)
                    all_results[data_ratio][embed_size][method]['wtau'].append(wtau)
                    all_results[data_ratio][embed_size][method]['scoring_time'].append(scoring_time)
                    all_results[data_ratio][embed_size][method]['training_time'].append(training_time)

            def avg_score(x, metric):
                try:
                    assert len(all_results[data_ratio][embed_size][x][metric]) == 8
                except: 
                    print(x, len(all_results[data_ratio][embed_size][x][metric]))
                    return "-1"
                return '%.2f'%np.mean(all_results[data_ratio][embed_size][x][metric])


            main_results[data_ratio][embed_size]['overall'] = pd.DataFrame({
                "Methods": map(lambda x: method_show_name[x], all_methods),
                "MRR": map(lambda x: avg_score(x, 'mrr'), all_methods),
                "Mean SCC": map(lambda x: avg_score(x, 'spr'), all_methods),
                "Training Time": map(lambda x: avg_time(all_results[data_ratio][embed_size][x]['training_time'], 'h'), all_methods),
                "Scoring Time": map(lambda x: avg_time(all_results[data_ratio][embed_size][x]['scoring_time'], 's'), all_methods)
            })

            # main_results[data_ratio][embed_size]['mrr'] = pd.DataFrame(np.array([[m] + ['%.3f'%s for s in all_results[data_ratio][embed_size][m]['mrr']] \
            #                                           + ['%.3f'%np.mean(all_results[data_ratio][embed_size][m]['mrr'])] for m in all_results[data_ratio][embed_size].keys()]), \
            #                                                     columns=['Methods'] + datasets +['MRR'])
            # main_results[data_ratio][embed_size]['spr'] = pd.DataFrame(np.array([[m] + ['%.3f'%s for s in all_results[data_ratio][embed_size][m]['spr']] \
            #                                           + ['%.3f'%np.mean(all_results[data_ratio][embed_size][m]['spr'])] for m in all_results[data_ratio][embed_size].keys()]), \
            #                                                     columns=['Methods'] + datasets +['Mean SC'])
    return all_results, main_results


data_ratios = [1.0]
embedding_sizes = [16, 32, 64, 128, 256, 384, 512, 768]
# embedding_sizes = [768]
all_results, main_results = summarize_results(data_ratios, embedding_sizes)
```

```python
plot_methods = main_results[1][768]['overall']['Methods'].tolist()
dims = list(main_results[1].keys())
X = list(main_results.keys())
Y = {}
# metric = 'Mean Spearman’s rank Correlation Coefficient'
metric = 'Mean SCC'
# metric = 'MRR'
for m in plot_methods:
    Y[m] = []
    for r in data_ratios:
        Y[m] += [float(main_results[r][768]['overall'].query(f'Methods=="{m}"')[metric].item())]
```

```python
# line_styles = [':'] * 2 + ['--'] * 2 + ['-.'] * 8 + ['-'] * 4
# marker_styles = ['d'] * 2 + ['^'] * 2 + ['*'] * 8 + ['>'] * 4

line_styles = ['-.'] * 8 + ['-'] * 4
marker_styles = ['*'] * 8 + ['>'] * 4

fig = plt.figure(figsize = (18,10), dpi=60)
ax = plt.axes()
ax.set_facecolor("aliceblue")
show_methods = [method_show_name[m] for m in static_methods + dynamic_methods]

for i, m in enumerate(show_methods):
    plt.plot(["%d"%(float(x)*100)+"%" for x in X], Y[m], label=m, mfc = "white", markeredgewidth=2, linestyle=line_styles[i], linewidth=3, marker=marker_styles[i], markersize=15)

plt.xlabel('Feature Dimensions', size=15)
plt.ylabel(metric, size=15)

plt.xticks(["%d"%(float(x)*100)+"%" for x in X], fontsize=15)
plt.yticks(fontsize=15)

plt.grid(zorder=0, linewidth=2, linestyle="-", color = "lavender")
plt.legend(loc="upper left", fontsize=16, ncol=8, bbox_to_anchor=(-0.02, 1.2))

plt.show()
```

```python
# find the best scores
# used_methods = [method_show_name[m] for m in static_methods + dynamic_methods]
used_methods = [method_show_name[m] for m in all_methods]

best_mrrs = []
best_sccs = []
best_ttimes = []
best_stimes = []
best_dims = []
for m in used_methods:
    mrr = 0
    scc = -1
    best_dim = 768
    for dim in embedding_sizes:
        if scc < float(main_results[1][dim]['overall'].query(f'Methods=="{m}"')['Mean SCC'].item()):
            mrr = float(main_results[1][dim]['overall'].query(f'Methods=="{m}"')['MRR'].item())
            scc = float(main_results[1][dim]['overall'].query(f'Methods=="{m}"')['Mean SCC'].item())
            ttime = main_results[1][dim]['overall'].query(f'Methods=="{m}"')['Training Time'].item()
            stime = main_results[1][dim]['overall'].query(f'Methods=="{m}"')['Scoring Time'].item()
            best_dim = dim
    best_mrrs += ["%.2f"%mrr]
    best_sccs += ["%.2f"%scc]
    best_ttimes += [ttime]
    best_stimes += [stime]
    best_dims += [best_dim]
best_results = pd.DataFrame({
    "Methods": used_methods,
    "MRR": best_mrrs,
    "Mean SCC": best_sccs,
    "Training Time": best_ttimes,
    "Scoring Time": best_stimes,
    "Best Dim": best_dims
})
best_dim_dict = dict(zip(best_results["Methods"], best_results["Best Dim"]))
```
